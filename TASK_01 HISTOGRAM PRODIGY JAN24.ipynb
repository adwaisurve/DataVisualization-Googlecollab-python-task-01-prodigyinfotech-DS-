```python
#load the dataset
data=pd.read_csv('orders.csv')
data.head(10)
```

```python
#  Bar chart based on the orders
state_counts = data['State'].value_counts()
colors = plt.cm.Paired(range(len(state_counts)))
plt.figure(figsize=(17, 8))
state_counts.plot(kind='bar', color=colors)
plt.title('Distribution of Orders by State')
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)

# colours
legend_labels = [plt.Rectangle((0,0),1,1, color=colors[i], edgecolor=None) for i in range(len(state_counts))]
plt.legend(legend_labels, state_counts.index, title='States', loc='upper right')

plt.show()
```
